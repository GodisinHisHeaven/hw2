```python
import torch

from torch.utils.data import DataLoader
from transformers import AdamW, AutoTokenizer, AutoModelForCausalLM

def load_base_model():
    tokenizer = AutoTokenizer.from_pretrained("Salesforce/codegen-2B-mono")
    model = AutoModelForCausalLM.from_pretrained("Salesforce/codegen-2B-mono", device_map='auto', torch_dtype=torch.float16)
    return model, tokenizer

model, tokenizer = load_base_model()
```

```python
import os
from tqdm.auto import tqdm
import re
from datetime import datetime
import ast

def save_bad_output(input_prompt: str, generated_code: str, function_name: str):
    """
    Saves bad model outputs for debugging.

    Args:
        input_prompt (str): The original prompt given to the model.
        generated_code (str): The raw generated output from CodeGen.
        function_name (str): The extracted function name (if any).
    """
    os.makedirs("debug_outputs", exist_ok=True)  # Create directory if not exists

    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    filename = f"debug_outputs/{function_name or 'unknown_function'}_{timestamp}.txt"

    with open(filename, "w", encoding="utf-8") as f:
        f.write("=== INPUT PROMPT ===\n")
        f.write(input_prompt + "\n\n")
        f.write("=== GENERATED CODE ===\n")
        f.write(generated_code + "\n\n")
        f.write("=== NOTES ===\n")
        f.write("Check if the function was not fully extracted or has extra/unwanted output.")

def extract_function(code: str, function_name: str) -> str:
    """
    Extracts exactly one function definition along with necessary imports.

    Args:
        code (str): The full generated code from CodeGen.
        function_name (str): The expected function name to extract.

    Returns:
        str: The extracted function with necessary imports.
    """
    import_lines = []
    function_code = []
    inside_function = False
    indentation_level = None

    # Split code into lines for structured parsing
    lines = code.split("\n")

    # Collect import statements at the beginning
    for line in lines:
        if line.startswith("import ") or line.startswith("from "):
            import_lines.append(line)
        elif line.strip():  # Stop collecting imports when reaching non-import lines
            break

    for line in lines:
        # Detect function definition
        if line.startswith(f"def {function_name}("):
            if inside_function:
                break  # Stop at the next function definition
            inside_function = True
            indentation_level = len(line) - len(line.lstrip())  # Determine function's base indentation
            function_code.append(line)
            continue

        # If inside function, capture lines with correct indentation
        if inside_function:
            current_indent = len(line) - len(line.lstrip())

            # Stop when indentation resets (new top-level function or class)
            if current_indent == 0 and line.strip() and not line.startswith(" "):
                break

            function_code.append(line)

    # If no function found, return empty string
    if not function_code:
        return ""

    # Combine imports with extracted function
    extracted_code = "\n".join(import_lines) + "\n\n" + "\n".join(function_code)

    # Validate with AST to ensure correctness
    try:
        ast.parse(extracted_code)
        return extracted_code
    except SyntaxError:
        return ""  # Return empty if invalid syntax


```

```python
def program_synthesis(input_prompt: str, model) -> str:
    """
    Generates a function using CodeGen 2B and extracts only the relevant function.

    Args:
        input_prompt (str): The function docstring and signature.
        model: The pretrained CodeGen model.

    Returns:
        str: The cleaned function implementation.
    """

    inputs = tokenizer(input_prompt, return_tensors="pt")
    input_ids = inputs["input_ids"].cuda()
    attention_mask = inputs["attention_mask"].cuda()
    with torch.no_grad():
        output_ids = model.generate(
            input_ids,
            attention_mask=attention_mask,
            max_length=512,
            temperature=0,
            pad_token_id=tokenizer.eos_token_id
        )

    generated_code = tokenizer.decode(output_ids[0], skip_special_tokens=True)

    # Extract function name from prompt
    function_name_match = re.search(r"def (\w+)\(", input_prompt)
    function_name = function_name_match.group(1) if function_name_match else ""

    # Extract and validate function
    extracted_function = extract_function(generated_code, function_name) if function_name else ""

    # Save bad outputs for debugging
    if not extracted_function:  # If extraction failed, log output for debugging
        save_bad_output(input_prompt, generated_code, function_name)

    return extracted_function


def complete_base_humaneval(model, dataset, workdir):
  for task_id, problem in tqdm(dataset.items()):
      name = task_id.replace("/", "_")

      prompt = problem['prompt']

      solution = program_synthesis(prompt, model)
      os.makedirs(os.path.join(workdir, name), exist_ok=True)

      with open(os.path.join(workdir, name, '0.py'), 'w') as f:
          f.write(solution)


# generate the solutions produced by codegen
complete_base_humaneval(model, dataset, "codegen_results")
```

```python
import json

def check_which_failed(workdir: str, dataset):
    with open(os.path.join(workdir, "eval_results.json"), "r") as f:
        results = json.loads(f.read())

    failed_humaneval = []
    failed_humaneval_plus = []

    for task_id in dataset.keys():
        total = results['eval'][task_id]['nfiles']
        humaneval_base = len([x for x in results['eval'][task_id]['base'] if x[0] == "success"]) / total
        humaneval_plus = len([x for x in results['eval'][task_id]['plus'] if x[0] == "success"]) / total

        if humaneval_base != 1:
            failed_humaneval.append(task_id)
        if humaneval_plus != 1:
            failed_humaneval_plus.append(task_id)

    return failed_humaneval, failed_humaneval_plus
```

```python
import re
import ast

def extract_function(code: str, function_name: str) -> str:
    """
    Extracts exactly one function definition along with necessary imports,
    even if imports are not at the beginning.

    Args:
        code (str): The full generated code from CodeGen.
        function_name (str): The expected function name to extract.

    Returns:
        str: The extracted function with necessary imports.
    """
    import_lines = []
    function_code = []
    inside_function = False
    indentation_level = None
    lines = code.split("\n")

    # Pass 1: Collect all import statements (regardless of their location)
    for line in lines:
        if line.startswith("import ") or line.startswith("from "):
            import_lines.append(line)

    # Pass 2: Extract exactly one function definition
    for line in lines:
        if line.startswith(f"def {function_name}("):
            if inside_function:
                break  # Stop at the next function definition
            inside_function = True
            indentation_level = len(line) - len(line.lstrip())  # Determine function's base indentation
            function_code.append(line)
            continue

        # If inside function, capture all lines with correct indentation
        if inside_function:
            current_indent = len(line) - len(line.lstrip())

            # Stop when indentation resets (new top-level function or class)
            if current_indent == 0 and line.strip() and not line.startswith(" "):
                break

            function_code.append(line)

    # Ensure a valid function was extracted
    if not function_code:
        return ""

    # Combine extracted imports with the function definition
    extracted_code = "\n".join(import_lines) + "\n\n" + "\n".join(function_code)

    # Validate with AST to ensure correctness
    try:
        ast.parse(extracted_code)  # Ensure it's valid Python
        return extracted_code
    except SyntaxError:
        return ""  # Return empty if invalid syntax
```

```python
def program_synthesis_improved(input_prompt: str, model) -> str:

    enhanced_prompt = f"""    # Rule:
# - Only generate one function.
# Now, complete this function:

{input_prompt}
"""

    inputs = tokenizer(enhanced_prompt, return_tensors="pt")
    input_ids = inputs["input_ids"].cuda()
    attention_mask = inputs["attention_mask"].cuda()
    with torch.no_grad():
        output_ids = model.generate(
            input_ids,
            attention_mask=attention_mask,
            max_length=1024,
            temperature=0,
            pad_token_id=tokenizer.eos_token_id
        )

    generated_code = tokenizer.decode(output_ids[0], skip_special_tokens=True)

    # Extract function name from prompt
    function_name_match = re.search(r"def (\w+)\(", input_prompt)
    function_name = function_name_match.group(1) if function_name_match else ""

    # Extract and validate function
    extracted_function = extract_function(generated_code, function_name) if function_name else ""

    # Save bad outputs for debugging
    if not extracted_function:  # If extraction failed, log output for debugging
        save_bad_output(enhanced_prompt, generated_code, function_name)

    return extracted_function


def complete_improve_humaneval(model, dataset, workdir):
  for task_id, problem in tqdm(dataset.items()):
      name = task_id.replace("/", "_")

      prompt = problem['prompt']

      solution = program_synthesis_improved(prompt, model)
      os.makedirs(os.path.join(workdir, name), exist_ok=True)

      with open(os.path.join(workdir, name, '0.py'), 'w') as f:
          f.write(solution)


# generate the solutions produced by codegen
complete_improve_humaneval(model, dataset, "codegen_results_improved")
```
