```python
import os
from tqdm.auto import tqdm
import re

def extract_function(code: str, function_name: str) -> str:
    """
    Extracts the function definition along with necessary imports.

    - Captures `import` and `from ... import ...` statements.
    - Extracts the function definition.
    - Removes any trailing explanations or irrelevant text.

    Args:
        code (str): The full generated code from CodeGen.
        function_name (str): The expected function name to extract.

    Returns:
        str: The extracted function along with necessary imports.
    """
    # Capture all imports at the beginning
    import_lines = []
    for line in code.split("\n"):
        if line.startswith("import ") or line.startswith("from "):
            import_lines.append(line)
        elif line.strip() == "":  # Stop when encountering an empty line
            break

    # Extract the function definition
    pattern = rf"(def {re.escape(function_name)}\(.*?\):.*?)(?=\n\n|\Z)"  # Match function block
    match = re.search(pattern, code, re.DOTALL)

    function_code = match.group(1).strip() if match else ""

    # Combine imports and function definition
    return "\n".join(import_lines) + "\n\n" + function_code if function_code else ""


def program_synthesis(input_prompt: str, model) -> str:
    """
    Generates a function using CodeGen 2B and extracts only the relevant function.

    Args:
        input_prompt (str): The function docstring and signature.
        model: The pretrained CodeGen model.

    Returns:
        str: The cleaned function implementation.
    """

    input_ids = tokenizer(input_prompt, return_tensors="pt").input_ids.to(model.device)

    with torch.no_grad():
        output_ids = model.generate(
            input_ids,
            max_length=512,  # Limit output length to reduce irrelevant text
            do_sample=False,  # Greedy decoding (no randomness)
            temperature=0,
            pad_token_id=tokenizer.eos_token_id
        )

    generated_code = tokenizer.decode(output_ids[0], skip_special_tokens=True)

    # Extract function name from prompt
    function_name_match = re.search(r"def (\w+)\(", input_prompt)
    function_name = function_name_match.group(1) if function_name_match else ""

    # Post-process to extract only the function
    return extract_function(generated_code, function_name) if function_name else ""


def complete_base_humaneval(model, dataset, workdir):
  for task_id, problem in tqdm(dataset.items()):
      name = task_id.replace("/", "_")

      prompt = problem['prompt']

      solution = program_synthesis(prompt, model)
      os.makedirs(os.path.join(workdir, name), exist_ok=True)

      with open(os.path.join(workdir, name, '0.py'), 'w') as f:
          f.write(solution)


# generate the solutions produced by codegen
complete_base_humaneval(model, dataset, "codegen_results")
```

```python
import json

def check_which_failed(workdir: str, dataset):
    with open(os.path.join(workdir, "eval_results.json"), "r") as f:
        results = json.loads(f.read())

    failed_humaneval = []
    failed_humaneval_plus = []

    for task_id in dataset.keys():
        total = results['eval'][task_id]['nfiles']
        humaneval_base = len([x for x in results['eval'][task_id]['base'] if x[0] == "success"]) / total
        humaneval_plus = len([x for x in results['eval'][task_id]['plus'] if x[0] == "success"]) / total

        if humaneval_base != 1:
            failed_humaneval.append(task_id)
        if humaneval_plus != 1:
            failed_humaneval_plus.append(task_id)

    return failed_humaneval, failed_humaneval_plus
```

```python
# you can use this to check which problem the model did not correctly solve
failed_humaneval, failed_humaneval_plus = check_which_failed("codegen_results", dataset)
```
